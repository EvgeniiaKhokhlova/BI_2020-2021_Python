```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from PIL import Image, ImageFilter, ImageDraw
from PIL.ImageFilter import (
   BLUR, CONTOUR, DETAIL, EDGE_ENHANCE, EDGE_ENHANCE_MORE,
   EMBOSS, FIND_EDGES, SMOOTH, SMOOTH_MORE, SHARPEN
)
from PIL.ImageFilter import (
    GaussianBlur
    )
from PIL.ImageFilter import (
    RankFilter, MedianFilter, MinFilter, MaxFilter
    )
```

```python
photo = Image.open('IMG_7231.JPG')
plt.imshow(photo)
```

```python
filter1 = photo.filter(FIND_EDGES)
plt.imshow(filter1)
```

```python
filter2 = photo.filter(GaussianBlur(radius=10))
plt.imshow(filter2)
```

```python
km = (
     -2, -1,  0,
     -1,  1,  1,
      0,  1,  2
      )
k = ImageFilter.Kernel(
    size=(3, 3),
    kernel=km,
    scale=sum(km), 
    offset=0  
    )
filter3 = photo.filter(k)
plt.imshow(filter3)
```

```python
filter4 = photo.filter(CONTOUR)
plt.imshow(filter4)
```

```python
filter5 = photo.filter(EMBOSS)
plt.imshow(filter5)
```
