```python
import matplotlib.pyplot as plt
import pandas
import seaborn as sns
import Bio
from Bio import SeqIO

plt.rcParams['figure.figsize']=(10,5) 

length_dist = map(len, Bio.SeqIO.parse('SRR.fastq', 'fastq'))
pandas.Series(length_dist).hist(color='salmon', bins=10)
plt.title('Sequence length distribution', fontsize=20)

plt.savefig('lenght_dist.png')
```
