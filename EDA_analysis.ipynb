```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("train.csv")
train
```

Firstly, lets look at the data structure and the number of NA:

Variable Descripton

PassengerId: Unique Id number of each passengers.
Survived : Whether the passenger survived(1) or died(0).
Pclass : Passenger class.
Name : Passenger name.
Sex : Passenger sex.
Age : Passenger age.
SibSp : Total number of siblings and spouses of the passenger have on ship.
Parch : Total number of parentss and children of the passenger have on ship.
Ticket : Ticket type.
Fare : Spent on ticket.
Cabin : Cabin category.
Embarked : Port where passenger embarked

```python
train.describe()
```

Next, we can look at the distrubution of variables

```python
plt.figure(figsize = (10, 5))

sns.countplot('Survived', data = train)
```

```python
We can see that more than half of the people have died.
```

```python
plt.figure(figsize = (10, 5))

sns.countplot('Sex', hue = 'Survived', data = train)
plt.show()
```

On this graph, we can see that there were more men than women and more women survived than men.

```python
plt.figure(figsize = (10, 5))

sns.countplot('Pclass', hue = 'Survived', data = train)
plt.show()
```

Passengers travelling in higher class survived more than passengers travelling in lower class. The main part of the passengers who died was in the 3rd class.

```python
figsize = (10, 5)
sns.catplot(x = 'Sex', y = 'Survived', data = train, kind = 'bar', col = 'Pclass')
plt.show()
```

But independent on ticket class women survived more often than a men.

```python
plt.figure(figsize = (10, 5))
plt.style.use('ggplot')

sns.distplot(train['Age'])
plt.show()
```

The middle age of passengers is about 30.

Let's look at age distribution whithin classes:

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.kdeplot(train['Age'][train['Pclass']==1], ax= ax)
sns.kdeplot(train['Age'][train['Pclass']==2], ax= ax)
sns.kdeplot(train['Age'][train['Pclass']==3], ax = ax)
plt.legend(['1', '2', '3'])
plt.show()
```

We see that more elderly people preferably choose a first-class tickets when younger people usually choose second or third class tickets.

```python
To check the correlation between variables we draw a heatmap:
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(train.corr(), annot=True, linewidths=0.05)
plt.show()
```

There are a weak correlation between Pclass/Fare because the 3rd class is cheaper, between Pclass/Survived because passengers of the 3rd class died more often, between Pclass/Age - older people have money for the 1st class.

Also, there is a positive correlation between Fare/Survived, which means that the more you paid for ticked the more you get a chance to survive.

So, the most important factors for survival were gender, class and money.
