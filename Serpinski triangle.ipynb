```python
import numpy as np
import pylab
from random import randint 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(7,5)

def midpoint(point1, point2):
    return [(point1[0] + point2[0])/2, (point1[1] + point2[1])/2]

random_point = [0,0] 

v1 = [0,0]
v2 = [1,0]
v3 = [.5, np.sqrt(3)/2]

for _ in range(5000):
    
    val = randint(0,2)
    if val == 0:
        random_point = midpoint(random_point, v1)
    if val == 1:
        random_point = midpoint(random_point, v2)
    if val == 2:
        random_point = midpoint(random_point, v3)
        
    plt.plot(random_point[0],random_point[1], 'k.', markersize=3)
    
plt.show()
```
