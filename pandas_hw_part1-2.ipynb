```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/Serfentum/bf_course/master/14.pandas/train.csv')
train
```

```python
plt.rcParams['figure.figsize']=(10,5) 
pos = list(train.iloc[:, list(train.columns).index('pos')])
nucleotides = train.filter(items=['A', 'T', 'G', 'C'])
nucleotides.index = pos
x_axis = nucleotides.plot(kind='bar', stacked=True)
x_axis.set_ylabel('Frequency distribution')
x_axis.set_xlabel('Position')
plt.savefig('nucleotide_frequency_distribution.png')
plt.show()
```

```python
mean_matches = train.loc[:, 'matches'].mean()
mean_matches_data = train.query('matches > @mean_matches')
train_part = mean_matches_data.filter(items=['pos', 'reads_all', 'mismatches', 'deletions', 'insertions'])
train_part.to_csv("train_part.csv")
```
